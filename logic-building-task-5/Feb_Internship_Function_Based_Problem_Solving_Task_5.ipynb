#Problem 1: Social Media – Post Engagement Analyzer

This function calculates total likes of a post and classifies whether the post is viral or has normal engagement based on the engagement threshold.

```python
def analyze_post(like_list):
  total_likes=0
  #loop to calculate total likes
  for like in like_list:
    total_likes+=like

  print("Total Likes:",total_likes)

  #check the likes condition
  if total_likes >= 1000:
    print("Post Status: Viral Post")
  else:
    print("Post Status: Normal Engagement")

# Get the number of likes from the user outside the function
user_likes = int(input("Enter the number of likes: "))

# Call the function, passing the single like as a list
analyze_post([user_likes])
```

#Problem 3: Agriculture – Rainfall Adequacy Checker

This function calculates the average seasonal rainfall and checks whether it meets the required adequacy level.

```python
def rainfall_check(rainfall_list, required_level):
  # Variable to store total rainfall
  total_rainfall=0

  # Loop to calculate total rainfall
  for rainfall in rainfall_list:
    total_rainfall+=rainfall

  # Calculate average rainfall
  average_rainfall=total_rainfall/len(rainfall_list)
  print("Average Rainfall:",average_rainfall)

  # Check if average rainfall meets required level
  if average_rainfall>=required_level:
    print("Rainfall Status: Adequate Rainfall")
  else:
    print("Rainfall Status: Inadequate Rainfall")

rain_data = [70, 75, 80, 63]
rainfall_check(rain_data, 70)
```

#Problem 4: Social Media – Duplicate Account Detection

This function detects duplicate usernames using set logic and flags if duplicate accounts exist.

```python
def check_duplicates(usernames):
    # Compare list length with set length.
    # If different, duplicates exist (set keeps only unique values).
    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


users = ["user1", "user2", "user1", "user3"]
check_duplicates(users)
```

#Problem 6: Agriculture – Premium Crop Price Filter

This function filters crop prices that exceed the premium pricing threshold.

```python
def premium_crops(prices):
    premium = []

    # Iterate through each crop price
    for price in prices:
        # If the price exceeds the premium threshold (2000), add it to the premium list
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)


crop_prices = [1500, 2500, 1800, 3200]
premium_crops(crop_prices)
```

#Problem 9: E-Learning – Student Attendance Eligibility System

This function calculates attendance percentage and determines eligibility for exams.

```python
def check_exam_eligibility(attendance_list):
    present = 0

    # Count the number of 'P' (present) entries in the attendance list
    for day in attendance_list:
        if day == "P":
            present += 1

    # Calculate the attendance percentage
    percentage = (present / len(attendance_list)) * 100
    print("Attendance Percentage:", percentage)

    # Determine eligibility based on a 75% attendance threshold
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


attendance = ["P", "P", "A", "P", "P"]
check_exam_eligibility(attendance)
```

#Problem 10: Smart Electricity Bill Analyzer

This function calculates electricity bills using slab-based pricing and classifies usage level.

```python
def calculate_bill(units):
    bill = 0

    # Calculate bill based on units consumed using slab-based pricing
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    # Classify usage based on the calculated bill amount
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    print("Total Bill:", bill)
    print("Usage Status:", status)


# Example usage of the function
calculate_bill(250)
```
