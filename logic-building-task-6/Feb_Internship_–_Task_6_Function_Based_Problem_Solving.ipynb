##Problem 1: Smart Parking Lot Management System

This function manages vehicle entry and exit logs, calculates current parked vehicles, tracks peak usage, and checks parking capacity.

```python
def manage_parking(capacity, logs):
    # Current parked vehicles
    parked = 0
    # Track peak parking usage
    peak_usage = 0

    # Process each log entry
    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        # Update peak usage
        if parked > peak_usage:
            peak_usage = parked

    print("Currently Parked Vehicles:", parked)

    # Capacity check
    if parked > capacity:
        print("Parking Status: Full (Over Capacity Alert)")
    else:
        print("Parking Status: Available")

    print("Peak Usage:", peak_usage)


manage_parking(50, ["IN","IN","IN","OUT","IN","IN","OUT"])
```

##Problem 2: Online Food Delivery Time Estimator

This function estimates delivery time based on distance, traffic conditions, and weather delays.

```python
def estimate_delivery_time(distance, traffic, weather):
      # 5 minutes per km
    base_time = distance * 5

    # Traffic delay
    if traffic.lower() == "high":
        base_time += 10
    elif traffic.lower() == "medium":
        base_time += 5

    # Weather delay
    if weather.lower() == "rainy":
        base_time += 5

    print("Estimated Delivery Time:", base_time, "minutes")


estimate_delivery_time(8, "High", "Rainy")
```

##Problem 3: Movie Theatre Seat Occupancy Analyzer

This function calculates occupancy percentage and determines show status such as Housefull or Almost Full.

```python
def analyze_seats(total_seats, booked_seats):

    # Calculate occupancy percentage
    occupancy = (len(booked_seats) / total_seats) * 100

    print("Occupancy:", occupancy, "%")

    # Determine show status
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


analyze_seats(200, [1]*150)
```

##Problem 4: Cloud Server Load Classification System

This function calculates average CPU usage and classifies server status as Normal, Warning, or Critical.

```python
def classify_server_load(cpu_readings):

    total = 0

    # Calculate total CPU usage
    for cpu in cpu_readings:
        total += cpu

    average = total / len(cpu_readings)
    print("Average CPU Load:", average, "%")

    # Classify load
    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


classify_server_load([45, 60, 70, 85, 90])
```

##Problem 5: Smart Classroom Resource Usage Monitor

This function tracks classroom resource usage and identifies overuse patterns with energy alerts.

```python
def monitor_resources(usage_dict):

    overused = []

    # Threshold assumed: > 8 hours = overuse
    for resource, hours in usage_dict.items():
        if hours > 8:
            overused.append(resource)

    print("Overused Resources:", ", ".join(overused) if overused else "None")

    if overused:
        print("Energy Alert: Yes")
    else:
        print("Energy Alert: No")


monitor_resources({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

##Problem 6: Online Event Registration Capacity Controller

This function manages event registrations, prevents overbooking, and activates waitlist mode when capacity is exceeded.

```python
def manage_registration(capacity, registrations):

    # Check if registrations exceed capacity
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


manage_registration(100, 105)
```
